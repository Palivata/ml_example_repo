```python
import sys
sys.path.append('d:\projects\hw1_modeling\.venv\lib\site-packages')
import pandas as pd
```

```python
data = pd.read_csv('../planet/train.csv')
```

```python
data['Id'] = data['image_name']
```

```python
data = data[['Id', 'tags']]
```

```python
all_labels = set()
for s in data['tags']:
    new_labels = s.split(' ')
    all_labels.update(new_labels)
```

```python
for label in all_labels:
    data[label] = 0
```

```python
for i, row in data.iterrows():
    for label in row['tags'].split(' '):
        data.loc[i, label] = 1
```
